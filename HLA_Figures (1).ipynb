```r
library(tidyverse)
library(bigrquery)

# This query represents dataset "hla-control" for domain "person" and was generated for All of Us Controlled Tier Dataset v8
dataset_52057127_person_sql <- paste("
    SELECT
        person.person_id,
        p_race_concept.concept_name as race,
        p_ethnicity_concept.concept_name as ethnicity,
        p_sex_at_birth_concept.concept_name as sex_at_birth 
    FROM
        `person` person 
    LEFT JOIN
        `concept` p_race_concept 
            ON person.race_concept_id = p_race_concept.concept_id 
    LEFT JOIN
        `concept` p_ethnicity_concept 
            ON person.ethnicity_concept_id = p_ethnicity_concept.concept_id 
    LEFT JOIN
        `concept` p_sex_at_birth_concept 
            ON person.sex_at_birth_concept_id = p_sex_at_birth_concept.concept_id  
    WHERE
        person.PERSON_ID IN (SELECT
            distinct person_id  
        FROM
            `cb_search_person` cb_search_person  
        WHERE
            cb_search_person.person_id IN (SELECT
                person_id 
            FROM
                `cb_search_person` p 
            WHERE
                has_lr_whole_genome_variant = 1 
            UNION
            DISTINCT SELECT
                person_id 
            FROM
                `cb_search_person` p 
            WHERE
                has_whole_genome_variant = 1 ) 
            AND cb_search_person.person_id IN (SELECT
                person_id 
            FROM
                `cb_search_person` p 
            WHERE
                has_ehr_data = 1 ) )", sep="")

# Formulate a Cloud Storage destination path for the data exported from BigQuery.
# NOTE: By default data exported multiple times on the same day will overwrite older copies.
#       But data exported on a different days will write to a new location so that historical
#       copies can be kept as the dataset definition is changed.
person_52057127_path <- file.path(
  Sys.getenv("WORKSPACE_BUCKET"),
  "bq_exports",
  Sys.getenv("OWNER_EMAIL"),
  strftime(lubridate::now(), "%Y%m%d"),  # Comment out this line if you want the export to always overwrite.
  "person_52057127",
  "person_52057127_*.csv")
message(str_glue('The data will be written to {person_52057127_path}. Use this path when reading ',
                 'the data into your notebooks in the future.'))

# Perform the query and export the dataset to Cloud Storage as CSV files.
# NOTE: You only need to run `bq_table_save` once. After that, you can
#       just read data from the CSVs in Cloud Storage.
bq_table_save(
  bq_dataset_query(Sys.getenv("WORKSPACE_CDR"), dataset_52057127_person_sql, billing = Sys.getenv("GOOGLE_PROJECT")),
  person_52057127_path,
  destination_format = "CSV")

```

```r
# Read the data directly from Cloud Storage into memory.
# NOTE: Alternatively you can `gsutil -m cp {person_52057127_path}` to copy these files
#       to the Jupyter disk.
read_bq_export_from_workspace_bucket <- function(export_path) {
  col_types <- cols(race = col_character(), ethnicity = col_character(), sex_at_birth = col_character())
  bind_rows(
    map(system2('gsutil', args = c('ls', export_path), stdout = TRUE, stderr = TRUE),
        function(csv) {
          message(str_glue('Loading {csv}.'))
          chunk <- read_csv(pipe(str_glue('gsutil cat {csv}')), col_types = col_types, show_col_types = FALSE)
          if (is.null(col_types)) {
            col_types <- spec(chunk)
          }
          chunk
        }))
}
control <- read_bq_export_from_workspace_bucket(person_52057127_path)

dim(control)

head(control, 5)
```

```r
notDiseased <- anti_join(control, HLAData, by = "person_id")
notDiseased
```

```r
not_diseased <- notDiseased %>%
count(race, name = "No")
not_diseased
```

```r
diseased <- HLAData %>%
count(race, name = "Yes")
diseased
```

```r
race_yes_no <- merge(diseased, not_diseased, by = "race")
race_yes_no
```

```r
race_counts <- race_yes_no %>%
  mutate(race = case_when(
    race %in% c("I prefer not to answer", "None Indicated", "PMI: Skip") ~ "Not Specified",
    TRUE ~ race
  )) %>%
  group_by(race)%>%
summarise(
    Yes = sum(Yes),
    No = sum(No),
    .groups = "drop"
  )
race_counts
```

```r
race_data <- race_counts %>%
  mutate(race = case_when(
    race %in% c("None of these", "More than one population") ~ "Other",
    TRUE ~ race
  )) %>%
  group_by(race)%>%
summarise(
    Yes = sum(Yes),
    No = sum(No),
    .groups = "drop"
  )
race_data
```

```r
race_long <- race_data %>%
  pivot_longer(cols = c("Yes", "No"), names_to = "Response", values_to = "Count")
race_long <- race_long %>%
  mutate(Log_Count = log10(Count + 1))
race_long
```

```r
race_colors <- c("#5CB4E4", "#CC78A9", "#D56327", "#0874B4", "#E7A125", "#1A9F74", "#b3de69", "#fccde5")
#race_colors <- c("#F7FBFF","#DEEBF7","#C6DBEF","#9ECAE1","#6BAED6","#4292C6","#2171B5","#084594")
race_levels <- unique(race_long$race)
race_color_map <- setNames(race_colors, race_levels)
graph2 <- ggplot(race_long, aes(x = Response, y = Log_Count, fill = race)) +
  geom_col(color = "black")+
geom_text(aes(label = Count), position = position_dodge(width = 0.9), angle = 90, vjust = -1, size = 2.3) +
facet_grid(rows = vars(str_wrap(race, width = 15)), switch = "y", space = "free")+
coord_flip()+
scale_y_reverse()+
scale_fill_manual(values = race_colors)+
scale_x_discrete(position = "top",labels = function(x) ifelse(x == "Yes", "**Yes**", x))+
labs(
y = "log(Number of participants)", x = "Has an HLA disease")+
theme(
    legend.position = "none",
    strip.placement = "top",
    strip.text = element_text(face = "bold", color = "white"),  
    strip.background = element_rect(fill = "dark grey", color = "black"),
    panel.background = element_rect(fill = NA, color = "black"),
    strip.text.y.left = element_text(angle = 0),
    axis.text.y = ggtext::element_markdown()
  )
graph2
```

```r
race_data <- race_long %>%
  group_by(race) %>%
  summarise(
    count = sum(Count[Response == "Yes"]),
    total = sum(Count),
    percentage = (count / total) * 100
  ) %>%
  arrange(percentage)

race_data$race <- factor(race_data$race, levels = race_data$race)

graph3 <- ggplot(race_data, aes(x = percentage, y = race, fill = race))+
  geom_col(color = "black")+ 
  geom_text(aes(label = paste("n =", count)),
            hjust = -0.1, 
            size = 3)+  
  scale_fill_manual(values = race_color_map)+
  scale_x_continuous(labels = scales::percent_format(scale = 1), limits = c(0, max(race_data$percentage) * 1.2))+
  labs(x = "Percentage with an HLA Disease", y = "Self-reported groups")+
  theme_minimal()+
  theme(
    legend.position = "none",
    panel.grid = element_blank(),
    panel.border = element_rect(color = "black", fill = NA, linewidth = 0.5),
    axis.line = element_blank(),
#     axis.text.y = element_text(face = "bold"),
    axis.ticks = element_line(color = "black")
  )+
  scale_y_discrete(labels = function(x) str_wrap(x, width = 15))
graph3
```
